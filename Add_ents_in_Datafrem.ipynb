```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
import numpy as np
import torch
import os
import pandas as pd
```

```python
df=pd.read_csv("/content/medium_Data.csv.xls")
df
```

```python
def datatoken(text):
  my_doc = nlp(text)
  token_list=[word for word in my_doc]
  # result=' '.join(token_list.text)
  
  return token_list
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
def removestopword(text): 
  sentence = nlp(text)
  notStopWords = [notStopWords.text for notStopWords in sentence if not notStopWords.is_stop]  
  joined = " ".join(notStopWords)
  return joined
```

```python
df=df.astype(str)
df=df.applymap(removestopword)
df
```

```python
import re
def removepunct(text):
  joined_filtered_sentence = re.sub(r'[^a-zA-Z0-9 ]',r'',text)
  return joined_filtered_sentence
```

```python
df['Titles']=df['Titles'].apply(removepunct)
df['Contents']=df['Contents'].apply(removepunct)
df
```

```python
def find_ORG_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'ORG']
  result=",".join(entities)
  return result

def find_PERSON_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'PERSON']
  result=",".join(entities)
  return result 
def find_DATE_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'DATE']
  result=",".join(entities)
  return result
def find_GPE_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'GPE']
  result=",".join(entities)
  return result
def find_PRODUCT_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if (i.label_ == 'PRODUCT') ]
  result=",".join(entities)
  return result    
```

```python
df.reset_index(level=0, inplace=True)
df['ORG'] = df['Contents'].apply(find_ORG_Ents)
df['PERSON']=df['Contents'].apply(find_PERSON_Ents)
df['GPE']=df['Contents'].apply(find_GPE_Ents)
df['PRODUCT']=df['Contents'].apply(find_PRODUCT_Ents)
df['DATE']=df['Contents'].apply(find_DATE_Ents)
df['ORG'] = df['ORG']+','+df['Titles'].apply(find_ORG_Ents)
df['PERSON']= df['PERSON']+','+df['Titles'].apply(find_PERSON_Ents)
df['GPE']= df['GPE']+','+ df['Titles'].apply(find_GPE_Ents)
df['PRODUCT']= df['PRODUCT']+','+df['Titles'].apply(find_PRODUCT_Ents)
df['DATE']= df['DATE']+','+df['Titles'].apply(find_DATE_Ents)
df
```
